```python
#Basic Imports
import pandas as pd
import warnings
warnings.filterwarnings('ignore')

# Web Scraping and parsing
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
# from driver import driver
```

```python
from constants import driver_path, brave_path
from selenium import webdriver
from selenium.webdriver.chrome.service import Service as BraveService
from webdriver_manager.chrome import ChromeDriverManager
from webdriver_manager.core.utils import ChromeType

# Define Brave path
options = webdriver.ChromeOptions()
options.binary_location = brave_path

# Create new automated instance of Brave
driver = webdriver.Chrome(service=BraveService(ChromeDriverManager(chrome_type=ChromeType.BRAVE).install()),options = options)
```

```python
## Getting Faculty list from the following links
faculty = ["https://www.iitg.ac.in/dsai/faculty", "https://www.iitg.ac.in/dsai/assoc_faculty_list", 
           "https://www.iitg.ac.in/dsai/distinguished_faculty"]

faculty_links = ["https://scholar.google.com/citations?user=a7LQA8cAAAAJ&hl=en&oi=ao",
                 "https://scholar.google.com/citations?user=0lsISW8AAAAJ&hl=en&oi=ao",
                 "https://scholar.google.com/citations?user=YwT89XIAAAAJ&hl=en&oi=ao",
                 "https://scholar.google.com/citations?user=j7oyJ0MAAAAJ&hl=en&oi=ao",
                 "https://scholar.google.com/citations?user=o4yGgBQAAAAJ&hl=en&oi=ao",
                 "https://scholar.google.com/citations?user=VtgWODQAAAAJ&hl=en&oi=ao",
                 "https://scholar.google.com/citations?user=aBIWxbEAAAAJ&hl=en&oi=ao",
                 "https://scholar.google.com/citations?user=oDcr8MIAAAAJ&hl=en&oi=ao",
                 "https://scholar.google.com/citations?user=2XR7_CsAAAAJ&hl=en&oi=ao",
                 "https://scholar.google.com/citations?user=W7nidBQAAAAJ&hl=en&oi=ao",
                 "https://scholar.google.com/citations?user=h8W9rUMAAAAJ&hl=en&oi=ao",
                 "https://scholar.google.com/citations?user=FYI_9IwAAAAJ&hl=en&oi=ao",
                 "https://scholar.google.com/citations?user=Es12tcUAAAAJ&hl=en&oi=ao",
                 "https://scholar.google.com/citations?user=XkrZYXsAAAAJ&hl=en&oi=ao",
                 "https://scholar.google.com/citations?user=L7yWFlkAAAAJ&hl=en&oi=ao",
                 "https://scholar.google.com/citations?user=sEltKKoAAAAJ&hl=en&oi=ao",
                 "https://scholar.google.com/citations?user=7bOSPf4AAAAJ&hl=en&oi=ao",
                 "https://scholar.google.com/citations?user=3qPiYJoAAAAJ&hl=en&oi=ao",
                 "https://scholar.google.com/citations?user=SverXLoAAAAJ&hl=en&oi=ao"]
faculty_links = [link+"&view_op=list_works&sortby=pubdate" for link in faculty_links]
```

```python
## Accessing the website
# link = "https://scholar.google.com/citations?hl=en&user=a7LQA8cAAAAJ&view_op=list_works&sortby=pubdate"
domain_data = pd.DataFrame()
for link in faculty_links:
    prof_data = pd.DataFrame(columns = ["Index", "All", "Since 2018"])
    driver.get(link)
    # Maximizing the window
    button = driver.find_elements(By.XPATH, '//*[@id="gsc_bpf_more"]')[0]
    button.click()
    driver.implicitly_wait(10)
    ## Accessing the domain names
    name = driver.find_elements(By.XPATH, '//*[@id="gsc_prf_in"]')[0].text
    domains = []
    elements = driver.find_elements(By.XPATH, '//*[@id="gsc_prf_int"]/a')
    for ele in elements:
        domains.append(ele.text)
    domain_data = domain_data.append([[name]+[domains]],ignore_index = True)
    cited_table = driver.find_elements(By.XPATH, '//*[@id="gsc_rsb_st"]')
    for row in cited_table[0].find_elements(By.XPATH, '//*[@id="gsc_rsb_st"]/tbody/tr'):
        prof_data = prof_data.append(pd.DataFrame([row.text.split(" ")], columns = ["Index", "All", "Since 2018"]), ignore_index=True)
    prof_data = prof_data.set_index('Index')
    prof_data.to_csv("prof_data/"+driver.find_elements(By.XPATH, '//*[@id="gsc_prf_in"]')[0].text.replace(" ", "_")+".csv")
domain_data.columns = ["Name", "Domains"]
domain_data.to_csv("prof_data/domain_data.csv", index = False)

# Manually add data for Arabin Kumar Dey: ["Distributions Models and Its Applications", "Statistics and Finance", 
# "Speech Signal Processing", "Machine Learning Algorithms"]

# Manually add data for Prabirkumar Saha: ['Process Engineering', 'Separation Process']
```

```python
for link in faculty_links:
    driver.get(link)
    # Maximizing the window
    button = driver.find_elements(By.XPATH, '//*[@id="gsc_bpf_more"]')[0]
    button.click()
    driver.implicitly_wait(10)
    name = driver.find_elements(By.XPATH, '//*[@id="gsc_prf_in"]')[0].text
    text_links = []
    links = []
    table_main = driver.find_elements(By.XPATH, '//*[@id="gsc_a_b"]/tr')
    for i in range(1, len(table_main)+1):
        text_links.append(table_main[i-1].find_elements(By.XPATH, f'//*[@id="gsc_a_b"]/tr[{i}]/td[1]/a')[0].get_attribute('text'))
        links.append(table_main[i-1].find_elements(By.XPATH, f'//*[@id="gsc_a_b"]/tr[{i}]/td[1]/a')[0].get_attribute('href'))
    publications_data = pd.DataFrame(columns = ['Title', 'Authors','Publication date','Journal','Volume','Issue','Publisher',
    'Description','Total citations','Scholar articles', 'Citation Count'])
    for i, link in enumerate(links):
        driver.get(link)
        keys = ['Title']
        keys = keys + driver.find_elements(By.CLASS_NAME, "gsc_oci_field")
        keys = [key if key == 'Title' else key.text for key in keys]
        values = [text_links[i]]
        values = values + driver.find_elements(By.CLASS_NAME, "gsc_oci_value")
        values = [value if value == text_links[i] else value.text for value in values]
        try:
            years = driver.find_elements(By.XPATH, '//*[@id="gsc_oci_graph_bars"]')[0].text.split("\n")
            citation_count = []
            for i in range(1, len(years)+1):
                citation_count.append(int(driver.find_elements(By.XPATH, f'//*[@id="gsc_oci_graph_bars"]/a[{i}]')[0].get_attribute('text')))
            keys.append("Citation Count")
            values.append(dict(zip(years, citation_count)))
        except:
            keys.append("Citation Count")
            values.append(dict())
        print(dict(zip(keys, values)))
        data = dict(zip(keys, values))
        try:
            data['Total citations'] = data['Total citations'].split('\n')[0].split(' ')[-1]
        except:
            data['Total citations'] = 0
        publications_data = publications_data.append(data, ignore_index=True) 
    publications_data.to_csv("publication_data/"+name.replace(" ", "_")+".csv", index = False)
```
